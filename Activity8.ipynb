<h1>Insight 1 -This heatmap highlights the areas in Chicago with the highest concentrations of criminal sexual assault incidents.</h1>

<h1>13. On which day does criminal damage happen the most, and where are the hotspots? </h1>
<h2>This map shows where criminal damage happens most in Chicago. It also tells which day has the most cases.</h2>

```python
damage_incidents = cc[cc['Primary Type'] == "CRIMINAL DAMAGE"]
damage_incidents['WeekDay'] = pd.to_datetime(damage_incidents['Date']).dt.day_name()
incidents_by_day = damage_incidents.groupby('WeekDay')['ID'].count().reset_index(name='event_count')
incidents_by_day = incidents_by_day.sort_values(by='event_count', ascending=False)
incidents_by_day

damage_data = damage_incidents.dropna(subset=['Longitude', 'Latitude'])

location_grouped = damage_data.groupby(['Longitude', 'Latitude']).size().reset_index(name='event_count')

location_grouped['weight_factor'] = (location_grouped['event_count'] - location_grouped['event_count'].min()) / \
                                    (location_grouped['event_count'].max() - location_grouped['event_count'].min())

heatmap_data = location_grouped[['Latitude', 'Longitude', 'weight_factor']].values.tolist()

map_center = [41.8781, -87.6298]
damage_map = folium.Map(location=map_center, zoom_start=10)

HeatMap(heatmap_data).add_to(damage_map)

damage_map.save('criminal_damage_updated_heatmap.html')

damage_map
```
