```python
import folium
from folium.plugins import HeatMap
```

```python
teeror = pd.read_csv('dataset\\Global_terrorist.csv', encoding='ISO-8859-1')
```

```python
teeror
```

```python
teeror=teeror[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude","location","summary","attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
teeror.head()
```

```python
teeror.isnull().sum()
```

```python
teeror["nkill"] = teeror["nkill"].fillna(0)
teeror["nwound"] = teeror["nwound"].fillna(0)
teeror["Casualty"] = teeror["nkill"] + teeror["nwound"]
```

```python
teeror.describe()
```

<h1>GEOSPATIAL WITH FOLIUM LIBRARY</h1>

```python
filtered_teeror = teeror[(teeror['region_txt'] == 'Middle East & North Africa') & (teeror['iyear'] >= 2000)]

# Aggregate incidents by location
aggregated_teeror = filtered_teeror.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

# # Normalize weights (optional)
aggregated_teeror['normalized_weight'] = (aggregated_teeror['incident_count'] - aggregated_teeror['incident_count'].min()) / \
                                (aggregated_teeror['incident_count'].max() - aggregated_teeror['incident_count'].min())

# Prepare heat_data
heat_data = aggregated_teeror[['latitude', 'longitude', 'normalized_weight']].values.tolist()

# Create a base map
m = folium.Map(location=[30, 40], zoom_start=4)

# Add heatmap layer
HeatMap(heat_data).add_to(m)

# Save or display the map
m.save('terrorism_heatmap.html')
m
```

<h1>insight1 -Region Analysis</h1>
<h2>Bombing/Explosion incidents in the Middle East & North Africa (2000+)</h2>
<h3>The heatmap visualization of Bombing/Explosion incidents in the Middle East & North Africa from the year 2000 onward, focusing on attacks that resulted in 10 or more fatalities, clearly highlights the regions most severely affected by violent terrorism. </h3>

```python
filtered_teeror = teeror[
    (teeror['region_txt'] == 'Middle East & North Africa') &
    (teeror['iyear'] >= 2000) &
    (teeror['attacktype1_txt'] == 'Bombing/Explosion') &
    (teeror['nkill'] >= 10)
]
agg_teeror = filtered_teeror.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
m1 = folium.Map(location=[30, 30], zoom_start=5)
HeatMap(data=agg_teeror[['latitude', 'longitude', 'incident_count']].values).add_to(m1)
m1
```

<h1>Insight 2 - Global Terrorism Heatmap by Casualties </h1>
<h3>This global heatmap visualizes terrorist incidents based on the number of casualties (deaths and injuries combined), revealing the geographic spread and intensity of terror attacks worldwide. High-casualty hotspots are visible in regions such as South Asia, the Middle East, and parts of Africa, reflecting the severe human impact of terrorism in these areas. </h3>

```python
teeror['Casualty'] = teeror[['nkill', 'nwound']].sum(axis=1)
heat_df = teeror.dropna(subset=['latitude', 'longitude', 'Casualty'])
heat_df = heat_df[heat_df['Casualty'] > 0]

m2 = folium.Map(location=[20,0], zoom_start=2)
HeatMap(heat_df[['latitude', 'longitude', 'Casualty']].values, radius=8).add_to(m2)
m2
```

<h1>Insight 3: TARGET ANALYSIS</h1>
<h2>This is a heatmap of terrorist attacks in South Asia from 2014 and up, where the target was Private Citizens & Property.</h2>
<h3>This heatmap highlights terrorist attacks targeting Private Citizens & Property in South Asia from 2014 onwards, emphasizing the regions where civilian life and infrastructure have been most affected. </h3>

```python
target_filtered = teeror[(teeror['iyear'] >= 2014) & (teeror['region_txt'] == 'South Asia')]
target_filtered = target_filtered[(target_filtered['targtype1_txt'] == 'Private Citizens & Property')]
aggregated_df = target_filtered.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
aggregated_df['normalized_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())
heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()
m = folium.Map(location=[30, 40], zoom_start=4)

HeatMap(heat_data).add_to(m)

m.save('terrorism_heatmap.html')
m
```

<h1>Insight 4: Hostage-Taking in Western Europe</h1>
<h3>This heatmap displays hostage-taking and kidnapping incidents across Western Europe, illustrating areas where this form of terrorism has been most active. The data reveals notable concentrations in countries like France, the UK, and Germany, indicating these regions have experienced multiple such events. </h3>

```python
Hostage1 = teeror[teeror['region_txt'] == 'Western Europe']
Hostage2 = Hostage1[Hostage1['attacktype1_txt'] == 'Hostage Taking (Kidnapping)']

location4 = Hostage2.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')
map_data4 = location4[['latitude', 'longitude', 'incident_count']].values.tolist()

map4 = folium.Map(location=[50, 10], zoom_start=4)
HeatMap(map_data4).add_to(map4)
map4
```

<h1>Insight5 -DEMOGRAPHIC ANALYSIS</h1>
<h2>This shows where Police Security Forces/Officers were attacked by terrorists in South America from 2000 and onwards.</h2>
<h3>This heatmap visualizes terrorist attacks targeting Police Security Forces and Officers in South America from the year 2000 onwards. The data reveals that specific areas in countries like Colombia and Peru experienced higher frequencies of these attacks.</h3>

```python
target_filtered = teeror[(teeror['iyear'] >= 2000) & (teeror['region_txt'] == 'South America')]
target_filtered = target_filtered[(target_filtered['attacktype1_txt'] == 'Police Security Forces/Officers')]
aggregated_df = target_filtered.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')
aggregated_df['normalized_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())
heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m = folium.Map(location=[30, 40], zoom_start=4)
HeatMap(heat_data).add_to(m)
m.save('terrorism_heatmap.html')
m
```
